# 태양광 발전량 예측 모델

- 2020년 1월 31일, 2020년 3월 31일, 2020년 5월 31일을 예측


- 하루동안의 데이터는 96개의 데이터가 생성


- 데이터는 15분 당 한개씩 생성


- 데이터 구조
    - 00:00:00 ~ 05:00:00 이 기간은 value=0 (20개) 
    - 05:15:00 ~ 20:15:00 이 기간은 value=? (63개)
    - 20:30:00 ~ 23:45:00 이 기간은 value=0 (13개)
    
- 총 63개의 각 시간을 맞추는 데이터를 분리하고 모델 생성

- 총 용량 133kw

### 1. 라이브러리 및 환경 세팅

```python
import pandas as pd
```

### 2. 데이터 불러오기

```python
data = pd.read_csv("./SolarPV_Elec_Problem.csv", header=None, names=["Date", "Value"])
```

```python
data.head()
```

### 3. 데이터 쪼개기

```python
def PSH_FUNCTION_1 (data, dataframe_new, time) : 
    
    for idx in data.index : 
    
        if time in str(data.loc[idx]["Date"]) : 

            sample = pd.Series({"Date" : data.loc[idx]["Date"], "Value" : data.loc[idx]["Value"]})

            dataframe_new = dataframe_new.append(sample, ignore_index=True)
            
    dataframe_new["past_01"] = dataframe_new["Value"].shift(1)
    dataframe_new["past_02"] = dataframe_new["Value"].shift(2)
    dataframe_new["past_03"] = dataframe_new["Value"].shift(3)
    dataframe_new["past_04"] = dataframe_new["Value"].shift(4)
    dataframe_new["past_05"] = dataframe_new["Value"].shift(5)
    dataframe_new["past_06"] = dataframe_new["Value"].shift(6)
    dataframe_new["past_07"] = dataframe_new["Value"].shift(7)

    dataframe_new = dataframe_new.dropna(subset=["past_01", "past_02", "past_03", "past_04", "past_05", "past_06", "past_07"])
    
    dataframe_new = dataframe_new.reset_index()
    
    dataframe_new = dataframe_new[dataframe_new.columns[1:]]
    
    return dataframe_new
```

```python
def PSH_FUNCTION_2 (dataframe, time) : 
    
    train = pd.DataFrame()
    test = pd.DataFrame()
    
    for idx in dataframe.index : 
        
        if dataframe.loc[idx]["Date"] == "2020-01-31T" + time + "+09:00" : 
            test = test.append(dataframe.loc[idx])
        elif dataframe.loc[idx]["Date"] == "2020-03-31T" + time + "+09:00" : 
            test = test.append(dataframe.loc[idx])
        elif dataframe.loc[idx]["Date"] == "2020-05-31T" + time + "+09:00" : 
            test = test.append(dataframe.loc[idx])
        else :
            train = train.append(dataframe.loc[idx])
            
    train = train.reset_index()
    test = test.reset_index()
    
    train = train[train.columns[1:]]
    test = test[test.columns[1:]]

    return train, test
```

```python
data_1430 = pd.DataFrame()
data_1430 = PSH_FUNCTION_1(data = data, dataframe_new = data_1430, time = "14:30:00")
data_1430_train = PSH_FUNCTION_2(data_1430, time="14:30:00")[0]
data_1430_test = PSH_FUNCTION_2(data_1430, time="14:30:00")[1]
```

```python
data_1430_train
```

```python
data_1430_test
```

### Train Valid Split & Linear Regression

```python
X = data_1430_train[data_1430_train.columns[2:]]
y = data_1430_train["Value"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.30, random_state=1234)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
reg.score(X_valid, y_valid)
```

```python
X_test = data_1430_test[data_1430_test.columns[2:]]
y_test = data_1430_test["Value"]
```

```python
reg.predict(X_test).round(4)
```

### LSTM 모델 구성

#### 2시 30분

```python
X_train = data_1430_train[data_1430_train.columns[2:]].values
X_test = data_1430_test[data_1430_test.columns[2:]].values

y_train = data_1430_train["Value"].values
y_test = data_1430_test["Value"].values

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 최종 트레이닝 셋
X_train_t = X_train.reshape(X_train.shape[0],7,1)
X_test_t = X_test.reshape(X_test.shape[0],7,1)

print("최종 Data")
print(X_train_t.shape)
print(X_test_t)
print(X_test_t.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.layers import LSTM
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping

K.clear_session()
model = Sequential()
model.add(LSTM(20,input_shape = (7,1)))
model.add(Dense(1))
model.compile(loss = 'mean_squared_error', optimizer = 'adam')
model.summary()
```

```python
model.fit(X_train_t,y_train, epochs = 100,
         batch_size = 16,verbose = 1)
```

```python
from keras.layers import LSTM
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping

K.clear_session()
model = Sequential()
model.add(LSTM(20,input_shape = (7,1)))
model.add(Dense(1))
model.compile(loss = 'mean_absolute_error', optimizer = 'adam')
model.summary()
```

```python
model.fit(X_train_t,y_train, epochs = 100,
         batch_size = 16,verbose = 1)
```

```python
y_pred = model.predict(X_test_t)
print(y_pred)
```
