### 3. 데이터 다듬기

```python
# 결측치 확인
plus_data.isnull().sum()
```

```python
# 강수, 일조, 일사, 적설은 결측치 0으로 채우기
values1=['Prec', 'Sunshine', 'Insolation', 'Snow']
for Value1 in values1:
  plus_data[Value1] = plus_data[Value1].fillna(0)

# 풍력, 풍향, 습도, 구름량, 가시거리, 지면온도는 결측치 평균으로 채우기
values2=['Wind_S', 'Wind_D', 'Humidity', 'Pressure', 'Cloud', 'Municipal', 'Ground temperature']
for Value2 in values2:
  plus_data[Value2] = plus_data[Value2].fillna(plus_data.mean()[Value2])
```

```python
# 결측치 재확인
plus_data.isnull().sum()
```

```python
plus_data.info()
```

```python
# 기존 데이터 시간대 변환
def utc_change (string) : 
    year = string[:4]
    month = string[5:7]
    day = string[8:10]
    hour = string[11:13]
    if hour in ":" :
        hour == int(string[11:12])
    else : 
        hour == int(string[11:13])
    minute = string[14:16]
    second = string[17:19]
    
    date = year + "-" + month + "-" + day + " " + str(hour)

    return date
utc_change(data.loc[0]["Date"])
data['Date'] = data['Date'].apply(lambda x :utc_change(x))
```

```python
# str데이터를 datetime으로 변환
data['Date'] = pd.to_datetime(data['Date'])
plus_data['Date'] = pd.to_datetime(plus_data['Date'])
```

### 1시간 단위로 변환 후 데이터 합치기

```python
# 15분 단위를 1시간 단위로
data2=data.resample('H', on='Date').sum()
```

```python
# data와 plus_data 병합
df = plus_data.assign(Value = Value.values)
df.head()
```

```python
corr = df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()


# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(16,10))
plt.title("Data Correlation", fontsize=30, position=(0.5, 1.0+0.05))
sns.heatmap(corr, annot=True, fmt='.2f', linewidths=.5, ax=ax, cmap='YlGnBu')
plt.show()
```

일조, 일사, 지면온도의 상관계수가 0.5 이상나오며 습도가 - 0.42로 가장낮은 음의 상관계수를 가진다.

빛의 강도가 가장 강한 시기는 봄과 가을이다. 여름도 발전량은 많지만 기상조건이 일정하지 않고 높은 온도로 인해 발전효율이 떨어진다(기술적 오류인데 기계가 25도에 최적화 되어있으므로 25도를 넘을 경우 효율이 떨어짐.
